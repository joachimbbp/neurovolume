# Importing test datasets

```python
from nilearn import datasets
```

```python
haxby_dataset = datasets.fetch_haxby() #2nd subject fetched by default
fmri_filename = haxby_dataset.func[0] #func is a list of filenames, one for each sbject
```

```python
print(fmri_filename)
```

# visualizing the fMRI volume
Data is 4D; it is a series of 3D EPI images

```python
import matplotlib
from nilearn import plotting
from nilearn.image import mean_img
```

```python
plotting.view_img(mean_img(fmri_filename), threshold=None)
```

No let's get these in matrices

The `Decoder` can translate Nifti images into matrices

```python
mask_filename = haxby_dataset.mask_vt[0]

plotting.plot_roi(mask_filename, bg_img=haxby_dataset.anat[0], cmap="Paired")
```
